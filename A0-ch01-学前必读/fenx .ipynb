```python
def getFileList(rootdir):   
    files = []
    for parent,dirnames,filenames in os.walk(rootdir):    # 三个参数：分别返回1.父目录 2.所有文件夹名字（不含路径） 3.所有文件名字    
        for filename in filenames:                        # 输出文件信息
            files.append(filename)                        # 输出文件路径信息       
    return files
```
