```python

import urllib2
from lxml import etree
import re
import pandas as pd
import numpy as np
url = "file:///C:/Users/kaiwei/Desktop/bsq/bsq2.htm"

txt = urllib2.urlopen(url).read()
res = etree.HTML(txt)
import warnings
warnings.filterwarnings("ignore")
# bf2 = res.xpath('/html/body/div[1]/div/table/tbody/tr[1]/td/p/span[1]/text()')[0].encode('utf-8')


# /html/body/div[1]
# /html/body/div[2]

# bf2 = res.xpath('/html/body/div//text()')[0].encode('utf-8')
data = open('datamm.csv','a')

df = pd.DataFrame(np.zeros((318,8)),columns=['company_name','company_address',
                                                  'company_desc','company_area',
                                                 'company_employee','sale2010',
                                                 'sale2011','sale2012'])

for i in range(0,318):
    try:
        df['company_name'][i] = res.xpath('/html/body/div['+str(i+2)+']/div[last()]/table/tbody/tr[1]/td/p/span//text()')
        df['company_address'][i] = res.xpath('/html/body/div['+str(i+2)+']/div[last()]/table/tbody/tr[2]/td/p/span//text()')

        df['company_area'][i] = res.xpath('/html/body/div['+str(i+2)+']/div[last()]/table/tbody/tr[9]/td[1]/p/span[2]/text()')
        df['company_desc'][i] = res.xpath('/html/body/div['+str(i+2)+']/div[last()]/table/tbody/tr[13]/td/p/span//text()')
        df['company_employee'][i] = res.xpath('/html/body/div['+str(i+2)+']/div[last()]/table/tbody/tr[26]/td[2]/p[3]/u/span/text()')

        df['sale2010'][i] = res.xpath('/html/body/div['+str(i+3)+']/div[1]/table/tbody/tr[9]/td[3]/p/span[1]/text()')
        df['sale2011'][i] = res.xpath('/html/body/div['+str(i+3)+']/div[1]/table/tbody/tr[10]/td[3]/p/span[1]/text()')
        df['sale2012'][i] = res.xpath('/html/body/div['+str(i+3)+']/div[1]/table/tbody/tr[11]/td[3]/p/span[1]/text()')
    except:
        pass
df
```
